```python
#asian population website
url='https://www.worldometers.info/population/countries-in-asia-by-population/'  
```

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
response=requests.get(url)
response
```

```python
response.text
```

```python
response.content
```

```python
soup=BeautifulSoup(response.content,'html.parser')
print(soup.prettify()) #to make above result beautiful
```

```python
title=soup.find("title")
title.text
```

```python
population_table=soup.find("tbody")
population_table
```

```python
body=population_table.find_all("tr")
```

```python
body_rows=body[0:26]
```

since there is no attributes/classes in other <td> i am going to take all <tr> as rows

```python
all_rows=[]
for row_num in range(len(body_rows)): # A row at a time
    row = [] # this will hold entries for one row
    for row_item in body_rows[row_num].find_all("td"): #loop through all row entries
        aa = row_item.text
        row.append(aa)
    all_rows.append(row[:8])
```

```python
row[:]
```

```python
print(all_rows)
```

```python
for i in all_rows:
    print(i)
```

```python
headings=['Rank','Country','Population(2020)','Yr_Change','Net_Change','Density(P/Km^2)','Land_area(km^2)','Migrants(net)']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=all_rows,columns=headings)
df
```

```python
df.to_csv('asian_countries_population.csv',index=False)
```
